```python
#jumlah total pinjaman sebelumnya (PREV_COUNT)
join_data = pd.merge(data, data_p.groupby('SK_ID_CURR').count()['SK_ID_PREV'], on='SK_ID_CURR', how='left')

#nilai paling umum NAME_CONTRACT_STATUS dari pinjaman sebelumnya
join_data = pd.merge(join_data, data_p.groupby(['SK_ID_CURR'])['NAME_CONTRACT_STATUS'].agg(lambda x: x.value_counts().index[0]), on='SK_ID_CURR', how='left')

#nilai maksimum positif dari DAYS_DECISION
join_data = pd.merge(join_data, -data_p.groupby('SK_ID_CURR')['DAYS_DECISION'].max(), on='SK_ID_CURR', how='left')
```

```python
#feature importance scoring by using Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(random_state=0, n_jobs=4)
forest.fit(X, y)
indices = np.argsort(forest.feature_importances_)[::-1]
cc = pd.DataFrame({'feature score':pd.Series(forest.feature_importances_),'features':pd.Series(X.columns)})   

#plot top 15 important features
plt.figure(figsize=(10,7))
sns.barplot(x='feature score',y='features',data=cc.sort_values(by='feature score',ascending=False).head(20))
plt.title('Feature importance score by Random Forest')
```

### Logistic Regression

```python
#select only top 50 important features
new_col = np.array(pd.Series(X.columns[indices]).head(50))
X_new = X[new_col]

#make pipeline with scaler
pipe = Pipeline([("scaler", StandardScaler()), ("logreg", LogisticRegression(random_state=0))])
param_grid = {'logreg__C': [10, 11, 12]}
fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# hyperparameters tuning
grid_logreg = GridSearchCV(pipe, param_grid=param_grid, cv=fold, scoring='roc_auc')
grid_logreg.fit(X_new,y)

print('Best hyperparameter: ', grid_logreg.best_params_)
print('Best cross validation score: ', grid_logreg.best_score_)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
fig, ax = plt.subplots(1,1,figsize=(6,6))
plot_confusion_matrix(grid_logreg, X_new, y, display_labels=['Not Risky','Risky'],cmap='Reds',ax=ax)
```
